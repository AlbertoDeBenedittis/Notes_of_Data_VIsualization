## Overview of plots in Matplotlib

__Plots__ in Matplotlib have a hierarchical structure. 
Each plot is encapsulated in a __Figure__ object.
This figure is the top-level container of the visualization. It can have multiple axes, which are basically individual plots inside this top-level container. <br>
Going a level deeper, we again find Python Objects that control axes, tick marks, legend,  title, textboxes, the grid, and many other objects. All these objects can be customized. 

The two main components of a plot are as follows:
* __FIGURE__: the figure is an outermost container and is used as a canvas to draw on it. It allows you to  draw multiple plots within it. It not only holds the Axes object but also the capability to configure the title. 
* __AXES__ : the axes in an actual plot, or subplot, depending on whether you want to plot single or multiple visualizations. Its sub-objects incldue the x and the y axis, spines and legends

### Anatomy of figure

* __Spines__ : Lines connecting the axis tick mark
* __Title__ : Text label of the whole figure object
* __Legend__ : Dscribe the content of the plot
* __Grid__ : Vertical and horizontal lines used as an extension of the tick marks
* __X/Y axis label___ : Text label for the X/Y axis below the spines
* __Minor Tick__ : Small value indicators between the major tick marks
* __Major Tick__ : Major value indicators on the spines
* __Major Tick Label__ : Text label that will be displayed at the major ticks
* __Line__: Plotting type that connects data point with a defined marker
* __Marker__ : Plotting type that plots every data point with a defined marker

#### Elegant solution

```python
days = ['Thur', 'Fri', 'Sat', 'Sun']
days_range = np.arange(len(days))
smoker = ['Yes', 'No']
bills
bills_by_days = [bills[bills['day'] == day] for day in days] 
```

What we have done in the above piece of code is the following:
1. We iterate along the days (with a list comprehension) 
2. we use the days as __filtering__ parameter
3. In this way we have obtained a list

```python
bills_by_days_smoker = [[bills_by_days[day][bills_by_days[day]['smoker'] == s] for s in smoker] for day in days_range]
```

In the above step we have performed a double list comprehension. 
What we have obtained is a list of 4 elements. The number of elements corresponds to the number of days considered in this analysis. These elements are themselves lists of length 2 which represent the classification Smoker Vs No-Smoker


```python
total_by_days_smoker = [[bills_by_days_smoker[day][s]['total_bill'].sum() for s in range(len(smoker))] for day in days_range]
```

In the above step we have perfromed a nested for in order to compute the sum of the elements according to the 2 conditions

```python
totals = np.asarray(total_by_days_smoker)
```

Here, the __asarray()__ function is used to convert any given input into an array. 

```python
totals
```

```python
# Create figure
plt.figure(figsize = (10,5), dpi = 300)
#Create stacked bar plot 
plt.bar(days_range, totals[:,0], label = 'Smoker')
plt.bar(days_range, totals[:,1], label = 'No-Smoker', bottom = totals[:,0])
# Add Legend
plt.legend()
# Add grid
plt.grid()
# Add title 
plt.title('Restaurant performance')
# Add labels
plt.xticks(days_range, days)
plt.ylabel('Daily total sales in $')
# Show plot 
plt.show()
```

#### Using a scatterplot to visualize correlation between various animals

__Note__: __Axes.set_xscale('log')__ and __Axes.set_yscale('log')__ change the scale of the x-axis and y-axis to a logarithmic scla, respectively.

```python
bio =  pd.read_csv('anage_data.csv')
```

```python
# Filtering according to Body Mass and Longevity
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
bio = bio[np.isfinite(bio[longevity]) & np.isfinite(bio[mass])]
```

```python
# Filtering data according to their class
amphibia = bio[bio['Class'] == 'Amphibia']
aves = bio[bio['Class'] == 'Aves']
mammalia = bio[bio['Class'] == 'Mammalia']
reptlia = bio[bio['Class'] == 'Reptilia']
```

```python
plt.scatter(longevity, mass, data = amphibia, label = 'Amphibia')
plt.scatter(longevity, mass, data = mammalia, label = 'Mammalia')
plt.scatter(longevity, mass, data = aves, label = 'Aves')
plt.scatter(longevity, mass, data = reptlia, label = 'Reptilia')
plt.legend()
# LOG SCALE
ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
# labels
plt.xlabel('Body mass in grams')
plt.ylabel('Maximum longevity in years')
plt.show()
```

### Layouts

There are multiple ways to define a visualization layout in Matplotlib. We can start by using __subplots__ and how to use the __tight layout__ to create visually appealing plots and then we will see __GridSpec__, which offers a more flexible way to create multi-plots.

#### Subplots

* __plt.subplots(nrows, ncols)__ creates a Figure and a set of subplots
* __plt.subplot(nrows, ncols, index)__ or equivalently __plt.subplot(pos)__ adds a subplot to the current figure. The index starts at 1.  __plt.subplot(2,1,1)___ is equivalent to __plt.subplot(211)___.
* __Figure.subplots(nrows, ncols)__ adds a set of subplots to the specific figure
* __Figure.add_subplot(nrows, ncols, index)__ or equivalently __Figure.add_subplot(pos)__ adds a subplot to the specified figure. 

For sharing the x or the y axis, the parameters __sharex__ and __sharey__ must be set, respectively.

### Radar charts 

Radar charts, also known as _spider_ or _web charts_, visualize multiple variables, with each variable plotted on ints own axis, resulting in a polygon. All axes are arranged radially, starting at the center with equal distance between each other and have the same scale. 

```python
data = pd.DataFrame({'Employee' : ['A','B','C','D'],
                    'Efficiency' : [5,4,4,3],
                    'Quality' : [5,5,3,3],
                    'Commitment' : [5,4,4,4],
                    'Responsible Conduct' : [4,4,4,3],
                    'Cooperation' : [4,3,4,5]})
data
```

```python
# Create angle values and close the plot
attributes = list(data.columns[1:])
values = list(data.values[:,1:])
employees = list(data.values[:, 0])
angles = [n / float(len(attributes))*2*np.pi for n in range(len(attributes))]
# Close the plot 
angles += angles[:1]
values = np.asarray(values)
values = np.concatenate([values, values[:, 0:1]], axis = 1)
```

```python
# Create subplots with the polar projection. Set a tught layout so that nothing overlaps:
plt.figure(figsize = (8,8), dpi = 150)
for i in range(4):
    ax = plt.subplot(2,2,i+1, polar = True)
    ax.plot(angles, values[i])
    ax.set_yticks([1,2,3,4,5])
    ax.set_xticks(angles)
    ax.set_xticklabels(attributes)
    ax.set_title(employees[i], fontsize = 14, color = 'r')

plt.tight_layout()
plt.show()
```

## Images

In this section, we will show you how to __load__, __save__, and __plot__ images with Matplotlib. 


### Basic Image Operations

* __Loading Images__
In case you encouter image formats that are not supported by MatPlotlib, we reccomand using the __Pillow__ library for loading the image. In Matplotlib, loading images is part of the image submodule. We use the alias __mpimpg__ for the submodule, as follows : `import matplotlib.image as mpimg`. 

`mpimg.imread(fname)` reads an image and returns it as a __numpy.array__. For grayscale images, the returned array has a shape (height, width), for RGB images (height, width,3)m and for RGBA (height, width, 4). The aaray values range from 0 to 255.

* __Saving Images__
`mpimg.imsave(fname, array)` saves a __numpy.array__ as an image file. If the __format__ parameter is not given, the format is deduced from the filename extension. With the optional parameters __vmin__ and __vmax__, the color limits can be set manually. For a grayscale image, the default for the optional parameter __cmap__ is _viridis_; you might want to change it to _gray_.


* __Plotitng a single Image__
`plt.imshow(img)` displays an image and returns an __AxesImage__. For grayscale images with shape(height, width), the image array is visualized using a colormap. The default colormap is _viridis_. 
For actually visualizing a grayscale image, the colormap has to be set to _gray_, that is `plt.imshow(img, cmap = 'gray')`.
Values for grayscale, RGB and RGBA images can be either _float_ ot _unit8_, aqnd range from __[0...1]__ or __[0...255]__, respectively. To manually define the values range, the parameters __vmin__ and __vmax__ must be specified. 

```python
import matplotlib.image as mpimg
```

```python
img_bici = mpimg.imread('bici.jpg')
```

```python
plt.imshow(img_bici)
plt.colorbar()
plt.show()
```

## Writing Mathematical Expression 

In case you need to write mathematical expression within the code, Matplotlib supports __TeX__. You can useit in any text by placing your mathematical expression in a pair of dollar signs. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0, (np.pi/4) * 7 , 8)  # left end = 0   right end = 7/4 pi   8 points
ys = np.sin(xs) + 3   # notice we know operate on arrays. All numpy functions can operate on them

plt.plot(xs, ys)

plt.title("c) solution     y = sin(x) + 3  with numpy and pi xlabels")
plt.xlabel(('$x$'))
plt.ylabel('$\sin(x)$')
# FIRST NEEDS A SEQUENCE WITH THE POSITIONS, THEN A SEQUENCE OF SAME LENGTH WITH LABELS
plt.xticks(xs, ["%sπ/4" % x for x in range(8) ])
plt.show()
```
