## Introduction

Seaborn is built on top of Matplotlib and provides a higher-level abstraction to make visually appealing visualizations.
A neat feature of Seaborn is the ability to integrate with DataFrames from pandas library. 
Internally Seaborn operates in DataFrames  and arrays that contain the complete dataset. This enables it to perform semantic mappings and statistical aggregations that are essential for displaying informative visualizations. Seaborn can also be solely used to change the style and apperance of Matplotlib visualization. 

The most prominent features of seaborn are as follows: 
* Beautiful out of the box plots with different themes
* Built-in color palettes that can be used to reveal patterns in the dataset
* Dataset-oriented interface
* A high-level abstraction that still allows for complex visualizations. 

One of the biggest advantages of Seaborn compared to Matplotlib is that it allows to easily operate on DataFrames and full dataset arrays, which makes this process simpler.

## Controlling Figure Aesthetics

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.title('Matplotlib')
plt.show()
```

The above is a Matplotlib plot. To switch to the Seaborn defaults, simply call the `set()` function:

```python
sns.set()
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.title('Seaborn')
plt.show()
```

Seaborn categorizes Matplotlib's parameters into two groups. The first group contains parameters for the aesthetics of the plot, while the second group scales various elements of the figure so that it can be easily used in different contexts, such as visualizations that are used for presentations, posters and so on. 

### Seaborn Figure Styles 

To control the style, Seaborn provides two methods: `set_style(style, [rc]` and `axes_style(style, [rc])`. 

`seaborn.set_style(style, [rc])` sets the aesthetic style of the plots. 

__Parameters__:
* __style__: A dictionary of parameters or the name of the one of the following preconfigured sets: _darkgrid, whitegrid, dark, white, or ticks_.
* __rc__ (optional): Parameter mappings to override the values in the present Seaborn style dictionaries. 


```python
sns.set_style('whitegrid')
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

`seaborn.axes_style(style, [rc]` returns a parameter dictionary for the aesthetic style of the plots. The function can be used in a with statement to temporarily change the style parameters. 

__Parameters__:

* __style__: A dictionary of parameters or the name of one of the following preconfigured sets: _darkgrid, whitegrid, dark white or ticks_
* __rc__ (optional): Parameters mappings to override the values in the preset Seaborn style dictionaries. 

```python
sns.set()
plt.figure()
with sns.axes_style('dark'):
    plt.plot(x1, label = 'Group A')
    plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

#### Removing Axes Spines

Sometimes, it might be desirable to remove the top and right axes spines:
`seaborn.despine(fig = None, ax = None, top = True, right = True, left = False, bottom = False, offset = None, trim = False)`
removes the top and right spines from the plot.

```python
sns.set_style('white')
plt.figure()
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
sns.despine()
plt.legend()
plt.show()
```

#### Context 

A separate set of parameters controls the scale of the plot elements. This is a handy way to use the same code to create plots that are suited for use in contexts where larger or smaller plots are necessary. To control the context, two functions can be used. 
`seaborn.set_context(context, [front_scale], [rc])` sets the plotting context parameters. This does not change the overall style of the plot, but affects things such as the size of the labels, lines, and so on. The base context is a __notebook__, and the other contexts are __paper__, __talk__ and __poster__, which are versions of the __notebook__ parameters scale by 0.8, 2.3, and 1.6 respectively. 

Here are the parameters: 
* __context__: A dictionary of parameters or the name of one of the following preconfigured sets: paper, notebook, talk, or poster
* __font_scale__ (optional): A scaling factor to independently scale the size of the font elements
* __rc__ (optional): Parameter mappings to override the values in the present Seaborn context dictionaries. 


```python
sns.set_context('poster')
plt.figure()
plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')
sns.despine()
plt.legend()
plt.show()
```

## Color Palettes

Color is a very important factor for your visualization. Color can reveal patterns in the data if used effectively or hide patterns if used poorly. 
Seaborn makes it easy to select and used __color palettes__ that are suited to your task. The __color_palette()__ function provides an interface for many of the possible ways to generate colors.
`seaborn.color_palette([palette], [n_colors], [desat])` returns a list of colors, this defining a color palette. 
Here are the parameters: 
* __palette__ (optional): Name of palette or None to return the current palette.
* __n_colors__ (optional): Number of colors in the palette. If the specified number of colors is larger than the number of colors in the palette, the colors will be cycled. 
* __deat__ (optional): Proprotion to desaturate each color by. 

You can set the palette for all plots with `set_palette()`. This function accepts the same arguments as `color_palette()`. In the following sections.


Link to Seaborn documentation: https://seaborn.pydata.org/tutorial/color_palettes.html

### Categorical color palettes

### Sequential color palettes 

__Sequential color palettes__ are appropriate when the data ranges from relatively low or uninteresting values to relatively high or interesting values. The following code snippets, along with their respective outputs, give us a better insight into __Sequential color palettes__.

```python
custom_palette = sns.light_palette('green')
sns.palplot(custom_palette)
```

```python
custom_palette = sns.light_palette('green', reverse = True)
sns.palplot(custom_palette)
```

```python
custom_palette = sns.dark_palette('green')
sns.palplot(custom_palette)
```

### Diveriging color palettes

__Diverging color palettes__ are used for data that consists of a well-defined midpoint. Emphasis is being laid on both high and low values. For example, if you are plotting any population changes for a particular region from some baseline population, it is best to use diverging colormaps to show the relative increasde and decrease in the populations. The following code snippet and output provides a better understanding about diverging plot, wherein we use the __coolwarm__ template, which is built into Matplotlib. 

```python
custom_palette2 = sns.color_palette('coolwarm', 7)
sns.palplot(custom_palette2)
```

You can use `diverging_palette()` function to create custom diverging palettes. We can pass two __hues__ in degrees as parameters, along with the total number of palettes. The following code snippet and output provides better insight:

```python
custom_palette3 = sns.diverging_palette(440,40, n = 7)
sns.palplot(custom_palette3)
```

### Kernel density estumation 

It is often useful to visualize how variables of a dataset are distributed. Seaborn offers handy functions to examine univariate and bivariate distributions. One possible way to look at a univariate distribution in Seaborn is by using the `distplot()` function. This will draw a histogram and fit a __kernel density estimate (KDE)__. 

```python
x = np.random.normal(size = 50)
sns.distplot(x)
```

To just visualize the KDE, Seaborn provides the `kdeplot()` function:

```python
sns.kdeplot(x, shade = True)
```

### Plotting Bivariate Distributions

For visualizing __bivariate distributions__, we will introduce three different plots. The first two plots use the `jointplot()` function, which creates a multi-panel figure that shows both the joint relationship between both variables and the corresponding marginal distributions. 

A scatter plot shows each observation as points on the _x_ and _y_ axis. Adittionally, a histogram for each variable is shown:

```python
data = pd.read_csv('salary.csv')
data.head(10)
```

```python
sns.set_style('darkgrid')
sns.jointplot(x = 'Salary', y = 'Age', data = data)
```

It is also possible to use the KDE procedure to visualize bivariate distributions. The joint distribution is shown as a contour plot.

```python
sns.jointplot(data = data, x = 'Salary', y = 'Age', kind = 'kde', xlim = (0,500000), ylim = (0,100))
```

## Visualizing Pairwise relationships

For visualizing multiple pairwise bivariate distributions in a dataset, Seaborn offers the `pairplot()` function. This function creates a matrix where off-diagonal elements visualize the relationship between each pair of variables and the diagonal elements visualize the relationship between each pair of variables and the diagonal elements show the marginal distributions. 

```python
data = pd.read_csv('basic_details.csv')
data
```

```python
sns.set(style = 'ticks', color_codes = True)
g = sns.pairplot(data, hue = 'Groups')
```

The pair plot, also called a __correlogram__ shows: scatterplots for all variables pairs on the off-diagonal, while KDEs are shown on the diagonal. Groups are highlighted by different colors. 

## Violin Plots

A different approach to visualizing statistical measures is by using __violin plots__. They combine box plots with the kernel density estimation procedure. 
It provides a richer description of the variable's distribution. Additionally, the quartile and whisker values from the boxplot are shown inside the violin. 

```python
data = pd.read_csv('salary.csv')
plt.figure(figsize = (10,10))
sns.set_style('darkgrid')
sns.violinplot('Education', 'Salary', hue = 'Gender', data = data, split = True, cut = 0)
```

## Multi-Plots in Seaborn

In the previous topic, we introduced a multi-plot, namely the pair plot. In this topic, we want to talk about the different way to create flexible multi-plots.

### FacetGrid

The __FacetGrid__ is useful for visualizing a certain plot for multiple variables separately. A FacetGrid can be drawn with up to three dimensions: __row__, __col__, and __hue__. The first two have the obvious correspondence to the rows and columns of an array. The __hue__ is the third dimension and shown with different colors. The __FacetGrid__ class has to be initialized with a DataFrame and the names of the variables that will form the row, column, or hue dimension of the grid. These variables should be __categorical__ or __discrete__. 


`seaborn.FacetGird(data,row,col,hue,...)` initializes a multi-plot grid for plotting conditional relationships. 
Here are some interesting parameters:
* __data__: A tidy ('long-form') DataFrame where each column corresponds to a variable and each row corresponds to an observation. 
* __row, col, hue__: Variables that define subsets of the given data, which will be drawn on separate facets in the grid. 
* __sharex, sharey__ (optional): Share __x/y__ axes across rows/columns
* __height__ (optional): Height (in inches) of each facet

Initializing the grid does not draw anything on the yet. For visualizing data on this grid, the __FacetGrid.map()__ method has to be used. You can provide any plotting function and the name(s) of the variable(s) in the data frame to plot. 

`FacetGrid.map(func, *args, **kwargs)` applies a plotting function to each facet of the grid. 
Here are the parameters: 
* __func__: A plotting functions that takes data and keyword arguments
* __*args__: The column names in data that identify variables to plot. The data for each variable is passed to __func__ in the order the variables are specified in. 
* __**kwargs__ : Keyword arguments that are passed to the plotting function. 

```python
data = pd.read_csv('salary.csv')
data
```

```python
g = sns.FacetGrid(data, row = 'District')
g.map(plt.scatter, 'Salary', 'Age')
```

```python
data = pd.read_csv('salary.csv')
g = sns.FacetGrid(data, col = 'District')
g.map(plt.scatter, 'Salary', 'Age')
```

### Water Usage

```python
water = pd.read_csv('water_usage.csv')
water
```

```python
per_title = []
usage = water['Usage'].tolist()
tot_perc = sum(water['Percentage'])
perc = water['Percentage']
for i in range(len(usage)):
    calc = int(perc[i]/tot_perc * 100)
    perc_x_tit = usage[i] + ' ({percentage}%)'.format(percentage = calc)
    per_title.append(perc_x_tit)

water['labels'] = per_title
water
```

```python
colors = sns.light_palette('orange', 4)
squarify.plot(sizes = water['Percentage'] , label = water['labels'], color = colors)
plt.axis('off')
plt.show()
```
