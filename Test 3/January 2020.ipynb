__Describe the chrts rule for graph classification and show and comment one example for each category.__

In order to classify a graph we have to consider two main variables: the __tone__ and the __experience__.

The _Tone_ refers to subtle variations between two choices: __reading__ and __feeling__.

With _reading_ we mean a very neutral tone. In this case the objective is optimising the ease of the wiever's estimation, the efficacy of perceiving the data; facilitate the understanding at high degree of precision and details, there is no need to seduce the audience, no need of visuals stimulation, the graph is analytical and pragramatical. 

On the other hand we have the _feeling_, graphs with feeling as tone we have a visual look that attracts and also informs, the audience need to be engaged. Wit this kind of plot we want to evoke strong emotions, we have to encapuslate emotions in the display. 

Typical examples of reading graphs are the ones used in scientific papers, they are very precise, they are highly scientific and just for a restricted and expert audience who can read it. An example of this kind of graph could be the one that we obtain when we do a _Lasso_ regression and we want to see for which value of $\lambda$ (which rules the shrinkage parameter) the value of the coefficients is 0. Indeed, this kind of plot by itself does not tell much. A non expert reader would probably end his/her graph reading process in the perception step.
Typical examples of felling graphs are the ones that we can found in magazines and newspapers. Indeed, with this kind of graph the author wants to catch the attention of the reader and make him think about what he/she has just read. A nice example, is the revers bar plot  (the blood plot) used to represent us casualities in golf war. Here the reverse bars plus the red color have a drammatic effect and impress the reader. 



The other variable that we have to consider is the __Experience__.
Experience means: how will the data viz pratically operate as a means of communication?; Through what functional experience will understanding be achieved by the viewer? 

We have three different type of experience: _Explanatory_, _Exhibitory_, and _Exploratory_. 

_Explanatory_: the author provides the viewer with a visual portrayal of the subject's data, taking responsability to bring key insights at the surface, rather than leaving it to the viewer, attempting to assist with the viewer's progress of understanding as much as possible drawing out the meaning of the data. In explanatory graphs, we have to annotate and describe: using annotations, using colors to separate different features, use captions to outline key messages, self-explanatory charts without the need for in-person explanation. 
An example of explanatory graph is the one titled: 'Black students are underrepresented on Campus'. This graph has many elements that help the reader. 
We have a subtitle that clarifies the content even more. We have clear label names for both the axes. We have a dotted line that should represent the ideal world and it has its own annotation, then we have a black solid line that represent the actual trend also that has its own annotation, then we have a scatterplot where each dot represent a university. Some universities (dots) are highighted and they are also present with the name of the institutions. All these elements enable the read to have a complete understanding of the concepts and ideas that the author wanted to tell. 

 

_Exploratory_: visualizations differ from expalanatory in that htey are focused on helping the viewrs or more specifically in this case, the users disvocer and form their own interpretations. Almost universally, these types of works will be digital and interactive in nature. In this kind of graphs it is frequent to find questions marks that are emblematic of explanatory visualizations as they are associated with a visualiser helping a user question.
An example of this is the representation titled 'Wood relative dimensional change'. It was developed to allow users to explore different measures concering the dimensional changes of different wood speices, over time, across selected cities of the world. There are no captions or conclusions. There are no indications of what is significant or insignificant. 
It is also important to say that today making an explanatory representation is easier thanks to interactive plots online. For example, thanks to bokeh it is possible to make interactive plots that enables to add information that pops up when the reader moves the cursor on that elements, more over it is possible to highlight some elements and but on the back ground other, then we can also switch the type of representation and move from dots to line easily.
However, a possible drawback of to the success of an exploratory visualization's impact is the 'so what?' factor. 'Why do you want me to do with this project?', 'Why is this relevant?'. If these are the reactions you are getting from users, then there is a clear disconnection between the intentions of your project and the experience of those using it.  


_Exhibitory_:visualizations are charactherized by being neither explicitely explanatory nor functionally explanatory. With exhibitory visualization the viewrs have to work to interpret the meaning relying on their own capacity to perceive and transalte features of viusalization. Generally describe these visualizations as simply being visual displayes of the data and find the ellipssi is emblematic of exhibitory visualizations, as it represents the idea of visualizer leaving the viewer to finish the task of gaining understanding. 
Think of this type of experience in relation to exhibiting an artwork: it takes the interpretative capacity of the viewer to be able to understand the content of a display as well as the context of a display. In contrast to exploratory visualisation, for exhibitory pieces this is conducted just by looking and thinking. But like exploratory experiences, exhibitory projects rely entirely on the audience having the motivation and capacity to interpret.
An example is the forbes visualization of the world's 100 highest-paid athletes.This is an exhibitory piece because it leaves you to form your own interpretation about what you see presented. In this case, it is quite clearly about the lack of representation of any female athlete among these top 100 earners. It does not need to be spelt out any more explicitly with captions or comments. It speaks for itself.


To conclude we can also say that using these two vairables, tone and experience, it is possible to create a matrix where we consider all the possible scenarions and we can place our plot. 

__Describe the PCA algorithm, highlighting its pros and cons; find a case study where PCA is more appropriate than MDS.__

The PCA is probably the most famous algorithm used for dimensionality reduction. The main idea of PCA as for the other method of dimensionality reduction is to reduce the number the dimensionality retaining as mach variability as possible.
The idea behind the Principal Component Analysis (PCA) is to project orthogonally a dataset $X = [x1,...xn]$ of np dimensional points into a r-dimensional space with $r = min(n-1,p)$, so that in the new coordinates the projecteed point' variables are uncorrelated. 
The new coordinates are called _principal components_, and each component is defined by the rules: being orthogonal to the previous component having highest possible values. These novel coordinates form an orthogonal basis. 


The schema of the PCA is the follwoing:
* center data;
* comput covariance matrix;
* extract eigen-values(vectors;
* normalize eigenvectors;
* diagonalize covariance matrix with variance on diagonal;
* compute the proportions of variances
In this way we are able to compute the components: the 1st component is the one which explains the biggest amount of variance inside the dataset and it corresponde to the eigenvector associated to the largest eigenvalue. Geometrically, the first principal components corresponds to a line that passes through the multidimensional mean and minimizes the sum of squares of the distances of the points from the line. 

The algorithm of the pca can be done in three different ways according to the kind of data that we have:
* covariance, when we have data with similar scales;
* correlation, when we have data with different scales;
* singular values decomposition, general purpose procedure. 

A drawback of PCA is that PCA cannot caputre the small variation in the data, it is onlt able to detect bug changes in the variance. Moreover, pca can shows a relation that does not exist as in the case of the double swiss roll. 

MDS refers to a set of related ordination techniques used in information visualization, in particular to display the information contained in a distance matrix. It is a form of non-linear dimensionality reduction.
Thus, we can say the PCA should perform better in cases of linear dimensionality reduction, so when data are low dimensional and they require at max two-three dimension. 


An example where we can prefer PCA over MDS is the one of the USArrests data set. For each of the 50 states in the United States, the data set contains the number of arrestsper 100000 residents for each of three crimes: Assault, Murder, and Rape. We also record UrbanPop (the percent of the population in each state living in urban areas). The results of the PCA on this dataset show that the first loading vector places approximately equal weight on Assault, Murder, and Rape, with much less weight on UrbanPop. Hence this component roughly corresponds to a measure of overall rates of serious crimes. The second loading vector places most of its weight on UrbanPop and much less weight on the other three features. Hence, this component roughly corresponds to the level of urbanization of the state.
Overall, we see that the crime-related variables (Murder, Assault, and Rape)
are located close to each other, and that the UrbanPop variable is far from
the other three. This indicates that the crime-related variables are correlated
with each other—states with high murder rates tend to have high
assault and rape rates—and that the UrbanPop variable is less correlated
with the other three.
