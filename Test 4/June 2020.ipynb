__Describe the step "Editorial Choices" in the Stage 3 of the Data Visualization Workflow; then provide 3 data visualization examples (not those in the lecture slides) to highlight focus, framing and angle concepts, respectively.__

Editorial thinking is concerned with making informed judgements about the content you intend to include in your visualisation.
In visualization we need to make selections about what analysis are going to portray to our audience in order to satisfy any particular curiosity.

When we are doing editorial choices we need to focus on three particular perspectives: 
* The __Angle__: is the first aspect of editorial thinking and concerns the angle of analysis, the view of your data that you think will best support the understanding required for this subject. 
* The __Framing__: after the angle, the second editorial choice is the framing. It concerns the refinement of contents that have to be included in the analysis.
* The __Focus__: the third editorial perspectives that concerns which items of your data you might choose to emphasise, thus generating some level of focus for the attention of the viewer. This decision is not a function of filtering  that is the concern of your framing decisions. Focus is about subjectively choosing to contrast visually features of your display that you deem to be more important than others.
For example, imagine that we are working for the F1 team of ferrari and we want to track the progresses that the engeneering team is doing. In order to show the results we are asked to make a plot. 
We have many possibilities for the choice of the angle:
* Which is the maximum speed reached?
* Which is the average speed/ time needeed for one lap of the track ? 
* Which is the best time for one lap of the track? 
Let's say that we want to consider the averege time needed for one lap of the track. 
Moving on we have the framing. We have to consider which data do we want to include:
* the data that we have gathered in lab conditions?
* the data obtained with the best pilot? 
* an averege of the results of the two pilots? 
* the results obtained during a sunny test day? 
Let's say that we want to consider only data coming from a simulation.
Lastly, we have to decide the focus: 
Thus, we can for example represent our graph as a copy of the track and showing with colors the part of the track according to the speed. 
For example, in green the ones where the perfromance was good; yellow for performance ok; and red for performances under the aspecatives.

__Describe the MDS family of algorithms, and comment similarities and differences with PCA. Then explain the differences between Classical, Metric and Non-Metric MDS, and provide an example where only one of the three above choices is adequate.__

Multidimensional scaling is a family of algorithms visualizing the level of similarity of individuale cases of a dataset. In practice, mds finds an embedding of n objects into a r-dimensional euclidean space $R^n$ so to preserve as well as possible (a function of ) the distances between original objects. MDS can be distinguished in 3 kinds, depending on the objectivve function: 
* classical mds - the objective function is called strain and involves directly the original distances between objects. 


* metric mds - the objective function is called stress and involves a function of the original distances. 

Given a dimension $p$ and a monotone function $f$, metric mds aims at finding $X = (x_{i}, ..., x_{n})$ such that $||x_i - x_j|| = f(d_{ij})$ as close as possible, i.e minimising a chosen loss function (stress)

* non-metric mds - the original distances are dissimilarities in the item-item matrix and the Euclidean distances between items, and defines the location of each of them. 

When dissimilarities are known only by their rank order, and the spacing between succesively ranked dissimilarities is of no interest or is unavailable. Here $f$ is only implicitely defines as a regression curve, and only preserves the order of d, that is $f(d_{ij} < f(d_{kl})$ if $d_{ij} < d_{kl}$ thus only the order of d is needed, not the actual values. The most common algorithm is the Kruskal mds.

Examples: 
* classical mds - Eickman's colors: 14 colors, different only in hue (wavelength), pairwaise rated as similar in a 0-5 scale by 31 people. In this case we use the original distances which are provided by the 'similaity matrix'.
* metric mds - Sammon mapping: The method was proposed by John W. Sammon in 1969. It is considered a non-linear approach as the mapping cannot be represented as a linear combination of the original variables as possible in techniques such as principal component analysis, which also makes it more difficult to use for classification applications. Denote the distance between ith and jth objects in the original space by ${\displaystyle \scriptstyle d_{ij}^{*}}\scriptstyle d^{*}_{ij}$, and the distance between their projections by ${\displaystyle \scriptstyle d_{ij}^{}}\scriptstyle d^{}_{ij}$. Sammon's mapping aims to minimize the following error function, which is often referred to as Sammon's stress or Sammon's error:$E = \frac{1}{\sum\limits_{i<j}d^{*}_{ij}}\sum_{i<j}\frac{(d^{*}_{ij}-d_{ij})^2}{d^{*}_{ij}}.$  Sammon mapping preserves the small $d_{ij}$, giving them a greater degree of importance in the fitting procedure than for larger values of $d_{ij}$
* non-metric mds -  An example of non-metric mds occurs each time we have only dissimilarities as rank order and the spacing between succesively ranked dissimilarities is of no interest or unavailable. Thus we could use non-metric mds when we have data about dissimilarity rating between sports.  


Compared to PCA we can say that both the algorithms perform dimensionality reduction. Nevertheless, PCA tends to perform better in cases of linear dimensionality reduction, so when data are low dimensional and they require at max two-three dimension. On the other hand, MDS algorithm can be used also for  non-linear dimensionality reduction.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import json
import geoplotlib 
from geoplotlib.utils import read_csv
from geoplotlib.colors import ColorMap

```

```python
df = pd.read_csv('economist_data.csv')
```

```python
ds = df.set_index('Country')
```

```python
geoplotlib.geojson('custom.geo.json', color = 'red')
geoplotlib.show()
```

```python
Countries = df.Country.unique().tolist()
```

```python
df.HDI.max()
```

```python
new_ds = {}
with open ('custom.geo.json') as data: 
    dataset = json.load(data)
    new_ds = dataset
    
    for i in range(len(dataset['features'])):
        name = (dataset['features'][i]['properties']['name']) #0 and 1 are the indeces according to the nation 
        if name in Countries:
            HDI = (ds.loc[name]['HDI'])
            CPI = (ds.loc[name]['CPI'])
            HDI_Rank = (ds.loc[name]['HDI.Rank'])
            new_ds['features'][i]['properties'] = {'name':name, 'HDI':HDI, 'CPI':CPI, 'HDI_Rank':HDI_Rank}
        else:
            new_ds['features'][i]['properties'] = {'name':name, 'HDI':0, 'CPI':0, 'HDI_Rank':0}
        
```

```python
# Converting our measure into a color
cmap = ColorMap('YlOrRd', alpha = 255, levels = 10)
def get_color(properties):
    return cmap.to_color(properties['HDI'], maxvalue = 1, scale = 'lin')
```

```python
geoplotlib.geojson(new_ds, fill = True, color = get_color,
                   f_tooltip = lambda properties: ('State: '+ properties['name'] + '\n' 
                                                   +'HDI:' + str(properties['HDI']) ))
geoplotlib.geojson(new_ds, fill = False, color = [255,255,255,255])
geoplotlib.show()
```
