__Discuss in detail the issue of trustworthiness in all three moments of data processing, data representation and data presentation, providing graphical examples for some of the discussed points (different from those included in the lecture notes).__

A good data visualization has to be trustworthy. The reder should not be fooled by a plot. As a matter of fact there are many ways through which a data visualization can be misleading. 
Starting from the beginning, the data that we used must be reliable and they have to come from reliable sources if not properly collected by the author. Moreover, the author should always communicate the operation that he/she has done on the data he/she worked with. For example, how he dealt with missing values if he ommitted all that observation or if he has substitued them with the mean or the median of that variable. For example, if we consider an online survey we as author should be careful in verifying if people have submitted their answers just one otherwise we must delete duplicates that could be bias the analysis.


Moving on, the data representation has to be trustworthy. We know that there are many ways to fool the reader. Indeed, human eye can be easily fooled by many visual effects such as 3D charts (barchart or pie chart) where the perspectives misrepresent the data. Moreover, there are differences that are not perceived in the same way by the human eye. For instance, we are less able to understand the differences between the area of two circles compare to the ability of combaring the height of two bars. 
For example, we have to be careful when we plot choroplet map because the human eye tends to give more importance to larger areas, so you might want to normalize your data by dividing the map area-wise.


Lastly, we have to present our data viz in the right way in order to make it readable by the reader. Thus, we have to provide all th graphical elements and annotations that make clear what we are showing. The reader has to be able to understand clearly what we are presenting. So, we have to choose the right layout, the most appropriate color schema, and all the annotations on the axes, the title a legend that make clearly visible what has to be notice. For example, in temporal representation a good practice could be creating a gray box in order to highlight as specific time period that is particular interesting as it is made in the above plot. In that graph we can also appreciate a proper choice of palettes: indeed we have colors from light green to blue to represent positive values and colors from yellow to red for negative values. These choices help the reader to immediately understand the plot. 

__Describe the PCA algorithm, and comment similarities and differences with MDS, tSNE and UMAP. Then provide a detailed example where using PCA is more adequate than using the other three methods, motivating such choice.__

PCA - Principal Component Analysis; the main idea is to project orthogonally a dataset $X = {x_{1}, x_{n}}$ of n-p dimensional points into a r-dimensional space with $r = min(n-1,p)$, so that in the new coordinates the projected points' variables are uncorrelated. 
The new coordinates are called principal components, and each component is defined by the rules: 
* being orthogonal to the previous components
* having highest possible variance
the novel coordinates from an orhtogonal basis.

PCA- fitting a p-dimensional ellipsoid to the data; each axsi of the ellipsoid is a principal component. 

PCA's schema: 
1) Center data,
2) Compute covariance matrix, 
3) extract eigen-values/vectors,
4) normalize eigenvectors,
5) diagonalize covariance matrix with variance on diagonal,
6) compute proportions of variances

Obviously, the four methods have in common the scope: they are both used to perform dimensionality reduction but they have different application and algorithms
* PCA vs MDS: Multidimensional is a family of algorithms visualizing the level of similarity of individual cases of a dataset. MDS provides also solution to cases when the distances are not euclidea thanks to metric MDS and non-metric MDS. 


* PCA vs t-SNE: PCA targets dimensionality preserving a large pairwise distances in the map, but cannot catch the strucutre of the data. For example, in the case of the double swiss roll the PCA detects relationships that do not exist. t-SNE focuses on neighbors thus we construct the concept of neighbourhood. In t-SNE the objective is to properly define the distances among the observation. The core algorithm is to find the minimum of the Kullback-Leiber function.  PCA is deterministic; t-SNE is not. In PCA the correct answer to the question posed is guaranteed. t-SNE might have multiple minima that might lead to different solutions. This necessitates multiple runs as well as raises questions about the reproducibility of the results. t -SNE is not learning a function from the original space to the new (lower) dimensional one and that's a problem. On that matter, t-SNE is a non-parametric learning algorithm so approximating with parametric algorithm is an ill-posed problem. The embedding is learned by directly moving the data across the low dimensional space. That means one does not get an eigenvector or a similar construct to use in new data. In contrast, using PCA the eigenvectors offer a new axes system what can be directly used to project new data. -SNE solves a problem known as the crowding problem, effectively that somewhat similar points in higher dimension collapsing on top of each other in lower dimensions (more here). Now as you increase the dimensions used the crowding problem gets less severe ie. the problem you are trying to solve through the use of t-SNE gets attenuated. You can work around this issue but it is not trivial. Therefore if you need a k dimensional vector as the reduced set and k is not quite small the optimality of the produce solution is in question. PCA on the other hand offer always the k best linear combination in terms of variance explained 




* PCA vs UMAP: The underlying idea behid UMAP is to detect the manyfold, the geometrical structure where the points are leaving and try to approximate it and project it in a two/three dimensional space. 

PCA probably perform better than the above mentioned algorithm when the distances are euclidean
