## Describe the step "Editorial Choices" in the Stage 3 of the Data Visualization Workflow; then provide 3 data visualization examples (not those in the lecture slides) to highlight focus, framing and angle concepts, respectively.
[0-5 points]

The data Visualization Workflow can be mainly divided into four stages: 
1. Brief Formulation 
2. Working with data (e.g. Data Processing)
3. Editorial Thinking, which mainly concerns the editorial choices 
4. Developing a design solution, thus, the representation of data itself 

Formulating some editorial choices deals with defining what one is going to show to the general audience. In particular, the author has to decide the perspective according to which data is going to be represented (e.g, the angle, the framing etc. is the author going to provide some general information, or some specific knowledge? From which point of view? On what is one going to mainly focus?), namely to understand what is most relevant to be presented, according to one's own opinion and the one of the audience, as well as understanding what has to be excluded from the main representation. 
Moreover, one has to understand how perspective and focus can be highlighted, namely, through patterns, color, or even interctiveness. 

Thus, the keyword to describe the "Editorial choices" stage includes both providing *sufficient* and *exhaustive* information, as well as *useful and relevant* focus and framing. This compromise is reached by the author when showing his chosen angle, while balancing it with the expectations of the general audience. Hereby, **relevance** is defined. This of course translates into choices of colouring, patterns and shapes, and some further questions such as: how is the visualization going to be? Busy? colorful or just concise? Moreover, are annotations useful for the general purpose of the graph?

In order:
1. Understand the relationship of the data I want to show - Angle
2. What data is going to be excluded/included - Framing 
3. Which data provides most and least meaning - Focus 
4. Define the chart type (decision mainly led by the Angle) 
5. Define how to convey the message clearly - Colors, annotations, etc.

## Describe the differences between tSNE and UMAP algorithms in dimensionality reduction, and provide an example (different from those included in the lecture notes) highlighting such differences.  
[0-5 points] 

tSNE algorithm uses a gaussian distribution to weigh neighbors distances to be mapped to a lower dimensional space. Its main aim is to construct a map projecting the high-dimensional space into a lower-dimensional one. The main difference from similar algorithms is that tSNE core lies on the fact that the original high-dimensional space's distance is defined with a gaussian, while in the lower-dimensional space a t-student is used. Thus, one wants to compute the Kullback Leibler distance. The hyperparameter to be tuned is perplexity, which is the variance sigma(i) for the high-dimensional gaussian of one point. Since density is likely to vary across data points, no single sigma value can be used.

On the other hand, the UMAP algorithm applies a real transformation map: we can transform both points, and the space by using a subset of points to construct the space (training to construct the map) and a test set to be projected in that particular map. Since the main goal is to preserve the main structure of the mainfold (where data lies), and optimization weights follow a Bernoulli probability, then the ideal distance to be computed is cross-entropy. 
Hereby, there are several hyperparameters but the most important to be tuned are the following:
1. the number of neighbors : the larger, the less local information is available 
2. the min distance to use : the larger, the better points distribution, loosing some significance in the local structure
3. the metric to measure the distance: it can be personally implemented, or chosen according to the task at hand.

One main remark follows the fact that UMAP is better suited for maintaining the overall global strucutre, while emphatizing the local one as well. To determine connectedness, UMAP uses the extension of a radiu to each point. Of course, the larger the radius, the more the global structure is enhanced, while loosing some information locally. 
Thus, one main remark is that with tSNE one defines perplexity, and therefore how "large" local relationships should be. On the other hand, UMAP generally selects some neighbours, thus taking much more care over the global structure, which may help with very large datasets. 



##  Describe the MDS family of algorithms, and comment similarities and differences with PCA. Then explain the differences between Classical, Metric and Non-Metric MDS, and provide an example where only one of the three above choices is adequate.

[0-5 points]

The MultiDimensional Scaling family of algorithms represent a well-defined alternative to PCA, which helps visualizing the level of similarity of individual cases of a dataset by finding an embedding in a lower-dimensional euclidean space which preserves the original distances. 
There are three main algorithms :
1. Classical MDS
2. Metric MDS
3. Non-Metric MDS 

Firstly, an analysis of the similarities and differences between PCA and MDS methods.
To start with, one may highlight that PCA is a method, while MDS is a whole class of analysis, where PCA leads to the same (metric) MDS result only if euclidean distance is used. The main difference here is that PCA uses a centered covariance matrix ($\frac{1}{n}X^tX$), while MDS uses a Gram Matrix after double-centering distance matrices $||G-X^tX||$.
The main aim of PCA is to be able to get the most relevant principal components, which are the ones guaranteiing the preservations of maximal variance (thus, the main spread of observations). To do so, Two main methods can be used, the Eigenvalue decomposition or the Sincular value decomposition, which lead to the main result $\frac{D^tD}{n}$, the diagonal matrix containing the respective eigenvalue of $X^tX$ eigenvectors, where each eigenvalue is proportional to the amount of variance explained by that specific component. 
MDS puts in place a slightly different apprach: what is maximised, is not the variance spread but the scattering of projections and thus *Euclidean Distance* has to be maximised. In the general case, the dimensionality reduction algorithm is the same as for the PCA, where observations are projected orthogonally on a matrix with a different base, which has a lower dimension. However, the optimization problem is hereby different:
1. PCA:  argmin ${\frac{1}{n}$\sum$ ||x_i - (u_i^tx_i)u_i||^2$ \


2. MDS: argmin $ ||G-X^tX||^2$


In particular, MDS can be split into three main cases : 
1. The classical MDS, for which solutions are unique up to isometries: we suppose $d_i_j$ to be an euclidean distance s.t. $||x_i-x_j||=d_i_j$ , then vector X is *not* unique, as for any c belonging to R^q a solution exists. 
For instance, another another possible for DR is to perform an eigendecomposition of the Gram matrix $C = X_cX^_c = US^2U^t$: where  principal components are its eigenvectors scaled by the square roots of the respective eigenvalues.
$X = (I1_n1_n)X$, where 1_n is a nn matrix of ones. From this  $C=XX^$ is a Gram matrix of uncentered data, which can be directly centered so that row and col means are zero. Considering now  the previously. defined matrix D of pairwise Euclidean distances, this matrix be converted into C_c. On the overall, this means that starting from the matrix of pairwise Euclidean distances D PCA can be performed to get principal components.

If a different distance is used, then solutions are not unique and approximation only is possible. Therefore, two further alternatives have been introduced.

2. The Metric and Non-metric MDS, in which conditions are relaxed. Instead of minimising the distance itself, the optimization problem aims to minimise the stress function, such that $||x_i - x_j||^2 = f(d_i_j)$
    
    a. The Metric MDS' aim is to minimise a defined loss function L. One main example is Sammon mapping, which allows to preserve smaller distances as well rather than enhancing large distances only (thus preserves inter-distances and smaller dissimilarities).
    b. The Non-Metric MDS used when dissmilarities are known only due to their ranked order. Hereb, the function is only an approximation of the regression curve and preserves the order of d only, where $f(d_{ij} < f(d_{kl}) if d_{ij} < d_{kl}$
    The most common algorithm is the Kruskal Algorith.
